```python
from matplotlib import pyplot as plt

# 2D vector
x = [4.5,8]
y = [10,8]

plt.quiver([0,0],[0,0],x,y,angles='xy',scale_units='xy',scale=2)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()


# 3D vector
x = [2.5,2.5,2.5]
y = [0,5.4,9]
z = [3.1,2.8,11.3]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z, 'blue')
plt.show()
```
